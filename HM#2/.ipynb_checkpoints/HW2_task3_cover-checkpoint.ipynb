### Data normalization (Pima Indians Diabetes Database)


Load the appropriate libraries.

```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from scipy import stats
from sklearn import preprocessing
% matplotlib inline
```

__3.1.__ Download the dataset from Kaggle [__(dataset)__](https://www.kaggle.com/uciml/pima-indians-diabetes-database#diabetes.csv).

```python
df = pd.read_csv('diabetes.csv')
df.head(20)
```

__3.2.__ Investigate the attributes and propose which of them need to be normalized? Explain your choice.

As for me, it looks like the insulin and glucose columns are  bit out of scale. And they may play less important role in the prediction.

```python
df.max()
```

Decimal Scaling Method For Normalization:

```python
df['Insulin'] = df['Insulin']/10
df.head(5)
```

__3.3.__ Perform normalization using the most appropriate method. Why did you choose it? How can you check
that it works better than others?

For this dataset I think that it would be best to use the min max normalization. Most of the data is in range 0 - 100 so i think it makes sense to use minMax normalization. 

```python
data = df.values 
minMaxScaler = preprocessing.MinMaxScaler()
scaledData = minMaxScaler.fit_transform(data)
df = pd.DataFrame(scaledData)
df.columns = ['Pregnancies','Glucose','BloodPressure',
              'SkinThickness','Insulin','BMI',
              'DiabetesPedigreeFunction','Age','Outcome']
```

```python
df.head()
```

__3.4.__ Why do we need to normalize numerical attributes?

It helps get more iformation from the data, for example when we have a coulumn that has values from 5000 to 5100, the diference between the min value and the maximum values is not so great, so this column will affect less the outcome then other columns while in the same time it can play a vital role in the outcome. So the data normalization helps you avoid this data ignorance so to say.

__3.5.__ What is standartization and how it differs from normalization?

The main purpose of the normalization is to bring the data into a range of 0 - 1. The standartization purpose is to transform the data into a way that it has a mean of 0 and a standard deviation of 1.
